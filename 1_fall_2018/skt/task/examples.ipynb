```python
import wolframalpha as wa
client=wa.Client(APPID)
```

```python
formulaes=[
    'x^2/4-y^2/9+z^2=1',
    'z=2-x^2/9-y',
    'z=5+sqrt(5-x^2-y^2+4x)',
    'z=x^2+1',
    'x^2-4x=y',
    '16x^2+y^2+4z^2=64',
    'x*y=1',
    'x^2/4-y^2/9-z^2=1',
    '3z-4y=12',
    'x^2-y^2=z',
]
results=[   
]

for a in formulaes:
    answer=client.query(a)
    results.append(answer)
```

```python
from IPython.display import Image, display  
import requests
from io import BytesIO

# Function to get image from API answer

def getimg(url):
    response = requests.get(url)
    data=BytesIO(response.content)
    return Image(data.read())
```

```python
i=0
for res in results:
    i=i+1
    print("## number %d" % i)
    for pod in res.pods:
        for subpod in pod.subpods:
            for img in subpod.img:
                display(getimg(img.src))
```
