### Задача 1
На шахматной доске $8\times8$ , у которой обрезаны четыре угловых клетки (т. е. всего 60 клеток), случайным образом расположены 8 ладей. Какова вероятность, что никаких пара этих ладей не бьет друг друга? Не забудьте описать вероятностное пространство.

**Решение***

Всего вариантов расположить $8$ ладей составляет $\binom{60}{8}$.

На любой полосе (вертикальной или горизонтальной) всегда должна быть только одна фигура.

Первую точку можно установить $60$ способами. 

Пронумеруем строки $\overline{0,7}$, также пронумеруем и столбцы $\overline{0,7}$.

Возникает отношение $(x_m,y_n): m,n \in \overline{0,7}$

Пусть $x$ пробегает $\overline{0,7}$, тогда в каждом наборе $\{(x,y)\}$ $y$ не должен повторяться.

Без ограничения общности ряды $1$ и $7$ можно обменять местами.

Тогда для ряда $0$ — $6$ возможностей расположить ладью, для ряда $1$ --- $5$, для ряда $2$: $8-2=6$.

\textbf{Получается ответ:} $6\cdot5\cdot6!/\binom{60}{8}$


### Задача 2
В круглую мишень радиуса $3$ наудачу производится три выстрела. Выигрыш при попадании в круг (с центром в центре мишени) радиуса $1$ равен $100$, а при попадании в круг радиуса $2$ --- $50$ (иначе выигрыш равен $0$). Найдите вероятность того, что в результате суммарный выигрыш не меньше $200$.

**Решение**

Будем считать, что нельзя попасть мимо мишени.

Посчитаем вероятности попадания в каждую из трех областей при одиночном броске:

(1) $\frac{\pi}{9\pi}=\frac{1}{9}$

(2) $\frac{3\pi}{9\pi}=\frac{1}{3}$

(3) $\frac{5\pi}{9\pi}=\frac{5}{9}$


Также будем считать, что все три выстрела производятся одновременно.



Результат $\ge200$ возможен только в случае, если (а) все три попадания в (1) или (б) одно в (1), а остальные в (2).
