## Задача 1

Джек выбирает университет, в котором бы он хотел получить высшее образование. Он
остановился на двух из них: 

 0. Гарварде
 1. Стэнфорде
 
и определил такие критерии выбора
университета: 

 0. размер стипендии
 1. престиж университета
 2. стоимость жизни
 3. достоинства города, где находится университет.  

---

 * Стипендия в Гарварде немного выше, чем в Стэнфорде.
 * Престиж обоих университетов примерно одинаков.
 * Стоимость жизни в Гарварде заметно дешевле,
 * Достоинства города, где расположен Стэнфорд заметно выше.

---

 * Достоинства города, где расположен университет, для Джека немного более важны, чем стоимость жизни в нем.
 * престиж университета немного важнее, по сравнению с городскими красотами. 
 * А вот размер стипендии значит гораздо больше даже по сравнению с престижем.

В какой университет вы посоветуете поступить Джеку? Чему равны средние рейтинги
университетов по критерию престижа? Чему равны средние веса критериев?


```python
import numpy as np
from numpy.linalg import norm

# Вводим матрицу важности критериев

crit = np.matrix([
    [1  ,7  ,1  ,1  ], # стипендия
    [1/7,1  ,1  ,3  ], # престиж
    [1  ,1  ,1  ,1/3], # стоимость жизни
    [1  ,1/3,3  ,1  ], # достоинства города
])

c=[

# 0. Стипендия в Гарварде немного выше, чем в Стендфорде
np.matrix([
    [1  ,3],
    [1/3,1]
]),

# 1. Престиж обоих университетов примерно одинаков.
np.matrix([
    [1,1],
    [1,1]
]),

# 2. Стоимость жизни в Гарварде заметно дешевле
np.matrix([
    [1  ,5],
    [1/5,1]
]),

# 3. Достоинства города, где расположен Стэнфорд заметно выше.
np.matrix([
    [1,1/5],
    [5,1  ]
]),
]

# Получаем относительные веса критериев
# деление элементов каждого столбца 
# на сумму элементов этого столбца
# искомые веса в виде средних значений 
# элементов строк нормализованной матрицы
crit = norm(crit, axis=1)


# Проверка согласованности
# Вычислить меру согласованности каждого критерия
# Определить индекс согласованности
# Вычислить коэффициент согласованности 
# как отношение ИС/ИР, где ИР - индекс рандомизации





harv=0
stan=0

for i in range(0,4):
    c[i]=norm(c[i], axis=1)
    harv=+c[i][0]*crit[i]
    stan=+c[i][1]*crit[i]
    #print(i)
    #print(c[i])
    #rint(crit[i])

print("Гарвард: %d, Стендфорд: %d" % (harv,stan))
```

Первый шаг процедуры МАИ состоит в попарном сравнении объектов по каждому критерию. 

Стандартная шкала сравнения:

| Рейтинг | Описание |
|--------|-------------|
| 1 | одинаковое предпочтение |
| 3 | умеренное |
| 5 | явное |
| 7 | очевидное |
| 9 | абсолютное |
